# Data cleaning
This data set is synthetic, generated from a small set (it seems only 30) of high-confidence "gold" labels.

## Restructing the dataset

I will "drop" the gold labels by inserting them as data, (probably) as intended during the generation of the data set. This way, I can (if desired) use a multi-task pipeline to do negative reason extraction followed by classification. 

I will also remove rows where their labels do not match the gold labels but have high confidence, or those that have a low sentiment confidence level, since these only serve as noise. I arbitrarily threshold this high confidence at 0.85.



```python
# ----- insert "gold" rows back into dataset -----

# find rows with gold sentiments or reasons
df_onlySentiment = df[(df["airline_sentiment_gold"].notnull()) 
                      & (df["negativereason_gold"].isnull())].copy()

df_onlyReasons = df[(df["airline_sentiment_gold"].isnull()) 
                    & (df["negativereason_gold"].notnull())].copy()

df_both = df[(df["airline_sentiment_gold"].notnull()) 
             & (df["negativereason_gold"].notnull())].copy()

# set relevant rows and confidence levels
df_onlySentiment["airline_sentiment"] = df_onlySentiment["airline_sentiment_gold"]
df_onlySentiment["airline_sentiment_confidence"] = 1

df_onlyReasons["negativereason"] = df_onlyReasons["negativereason_gold"]
df_onlyReasons["negativereason_confidence"] = 1

df_both["airline_sentiment"] = df_both["airline_sentiment_gold"]
df_both["airline_sentiment_confidence"] = 1
df_both["negativereason"] = df_both["negativereason_gold"]
df_both["negativereason_confidence"] = 1

# merge with the main dataframe and remove duplicates
df = pd.concat([df, df_onlySentiment, df_onlyReasons, df_both]).drop_duplicates()
```

```python
'''
This method takes two dataframes and performs a set difference (df1 - df2).
It returns the result as a dataframe.
'''
def set_difference(df1, df2):
    return pd.concat([df1, df2, df2]).drop_duplicates(keep = False)
```

```python
# remove noisy data
threshold = 0.90
df_wrongSentiment = df[(df["airline_sentiment"] != df["airline_sentiment_gold"]) 
                       & (df["negativereason"] == df["negativereason_gold"])
                       & df["airline_sentiment_confidence"] > threshold]

df_wrongReason = df[(df["negativereason"] != df["negativereason_gold"])
                   & (df["negativereason_confidence"] > threshold)
                   & (df["negativereason"].notnull())
                   & (df["negativereason_gold"].notnull())]

df_lowSentimentConfidence = df[df["sentiment_confidence"] < threshold]


df = set_difference(df, df_wrongSentiment)
df = set_difference(df, df_wrongReason)
df = set_difference(df, df_lowSentimentConfidence)
```

Now, we can select the columns of interest (to what would actually be seen at inference), and then encode relevant categorical variables into numerical ones so that we can explore the data. 

```python
# focus on informative columns
desired = ["airline_sentiment",
           "retweet_count",
           "text",
           "user_timezone"]
df = df[desired]
df.head()

# convert categorical columns to numbers
cat_columns = ['airline_sentiment',
               'retweet_count',
               'user_timezone']

df[cat_columns] = df[cat_columns].apply(lambda x : x.astype('category').cat.codes)
df.head()
```

## Exploring and extracting features

### Extracting features
There are a few heuristics we can use to engineer useful features. Negative tweets intuitively tend to be longer and have more capital letters on average (imagine someone screaming). I would also consider identifying hashtags, but this dataset does not have any.


```python
# methods to extract features of text
def extract_hashtags(text):
    import re
    return re.findall('(#[A-Za-z]+[A-Za-z0-9-_]+)', text)

def character_count(text):
    return len(text)

def word_count(text):
    return len(text.split())

def upperCase_count(text):
    return sum(1 for c in text if c.isupper())
```

```python
df['wordCount'] = df['text'].apply(lambda text: word_count(text))
df['upperCaseCount'] = df['text'].apply(lambda text: upperCase_count(text))
```

## Handling class imbalance
There are a few ways we can handle class imbalance. Currently, there are about twice as many negative classes as there are positive and neutral. A simple way is thus to oversample the positive and neutral classes, and undersample the negative ones such that they are equal. Other approaches like generating synthetic data are too complex, while pure undersampling will reduce the size of the dataset. 

There's a tradeoff between reducing the size of the dataset and increasing it (learning machine-generated noise), so I think an in-between is probably the way to go.

```python
n_ideal = len(df)//3

df_neg = df[df["airline_sentiment"] == 0].sample(n_ideal, replace = False)
df_neutral = df[df["airline_sentiment"] == 1].sample(n_ideal, replace = True)
df_pos = df[df["airline_sentiment"] == 2].sample(n_ideal, replace = True)

df_balanced = pd.concat([df_neg, df_neutral, df_pos])

# save
df_balanced.to_csv('cleanedAndBalanced.csv')
df_balanced.head()
```

## Text preprocessing
Before feeding our text into a model, we need to preprocess our text to remove words that are not very semantically meaningful for sentiment classification. These steps include removing "@" characters, stop words, etc. We do it here. 

```python
'''
This method takes in a string and cleans it.
It returns a string or optionally a list of string tokens.
'''
def clean_text(text, string = True):
    # ----- helper functions -----
    def clean_case(text):
        return text.lower()

    def clean_URL(text):
        import re
        return re.sub('((www.[^s]+)|(https?://[^s]+))', ' ', text)

    def clean_numbers(text):
        import re
        return re.sub('[0-9]+', '', text)

    def clean_AT_USER(text):
        import re
        return re.sub('@[^\s]+','', text)

    def clean_mentions(text):
        import re
        return re.sub(r'@[A-Za-z0-9_]+', '', text)

    def clean_hashtags(text):
        import re
        return re.sub(r'#[A-Z0-9]+', '', text)

    def clean_retweets(text):
        import re
        return re.sub(r'RT : ', '', text)

    def clean_newLines(text):
        import re
        return  re.sub(r'\n', ' ', text)

    def clean_htmlCharacters(text):
        import re
        return re.sub(r'\&\w*;', '', text)

    def clean_stopwords(text):
        from nltk.corpus import stopwords
        stopwords_list = stopwords.words('english')
        return " ".join([word for word in str(text).split() if word not in stopwords_list])

    def clean_punctuation(text, tokens = True):
        import string
        english_punctuation = string.punctuation
        translator = str.maketrans('', '', english_punctuation)
        if tokens:
            return [word.translate(translator) for word in text]
        else:
            return text.translate(translator)

    def clean_repeating_char(text):
        import re
        return re.sub(r'(.)1+', r'1', text)

    def clean_shortWords(text):
        return " ".join([word for word in str(text).split() if len(word) > 2])

    def tokenize(text):
        return text.split(' ')

    def lemmatizer(text):
        import nltk
        l = nltk.WordNetLemmatizer()
        text = [l.lemmatize(word) for word in text]
        return text

    def clean_emoji(text):
        import re
        regrex_pattern = re.compile(pattern = "["
            u"\U0001F600-\U0001F64F"  # emoticons
            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
            u"\U0001F680-\U0001F6FF"  # transport & map symbols
            u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               "]+", flags = re.UNICODE)
        return regrex_pattern.sub(r'',text)
    
    # ----- clean text -----
    text = clean_case(text)
    text = clean_AT_USER(text)
    text = clean_mentions(text)
    text = clean_hashtags(text)
    text = clean_retweets(text)
    text = clean_shortWords(text)
    text = clean_URL(text)
    text = clean_htmlCharacters(text)
    text = clean_newLines(text) 
    text = clean_emoji(text)
    text = clean_repeating_char(text)
    
    # remove sentiment-agonostic tokens
    text = clean_numbers(text)
    text = clean_stopwords(text)
    
    # tokenization
    text = tokenize(text)

    # lemmatization
    text = lemmatizer(text)

    # remove puncutation
    text = clean_punctuation(text)
    
    if string:
        text = "".join([word + " " for word in text])
    return text
```

## Logistic regression model
The logistic regression model is a simple, efficient and fast model that tends to work well for simple classification models. I did consider other models, and in particular found tree-based models like CatBoost interesting. But for simplicity, we can train a logistic regression model. Note that the classification labels are ordinal rather than one-hot-encoded since a logistic regression model is normally used for binary classification.

We do so by defining the model and then using k-fold validation to assess its performance.

```python
'''
This method encapsulates the evaluation of the logistic regression model.
It returns the results of recall, precision, f1 and accuracy.
'''
def evaluate(model, X, y):
    from sklearn.metrics import classification_report
    y_pred = model.predict(X)
    result = classification_report(y, y_pred)
    print(result)
    res = classification_report(y, y_pred, output_dict = True)
    result = res['weighted avg'].values()
    result = np.array(list(result))
    return np.append(result, res["accuracy"])
```

```python
'''
This method encapsulates the k-fold evaluation of the logistic regression model.
It displays the results of each run, all the runs on average and returns them as 
an array.
'''
def kfold_train(X, Y, n_splits):

    results = np.empty((n_splits,5), dtype = 'float64')
    count = 0       
                       
    # train test kfold
    from sklearn.model_selection import KFold
    kf = KFold(n_splits = n_splits, shuffle = True).split(X)
    for train_index, test_index in kf:
        print(f"\n##### NEW RUN {count + 1} ##### \n")
        
        # set train and test
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = Y.iloc[train_index], Y.iloc[test_index]
        
        
        # ----- define model ----- 
        ##### CHANGE HERE #####
        
        # encoding with n-grams by fitting on training data
        from sklearn.feature_extraction.text import TfidfVectorizer
        vectoriser = TfidfVectorizer(ngram_range=(1,2)) # use unigrams and bigrams
        vectoriser.fit(X_train)

        # transform into sparse vectors
        X_train, X_test = vectoriser.transform(X_train), vectoriser.transform(X_test)

        from sklearn.linear_model import LogisticRegression
        model = LogisticRegression(penalty = 'l2', C = 1e5, max_iter = 1000, n_jobs=-1)
        model.fit(X_train, y_train)
        
        # evaluation
        metrics_names = ["precision", "recall", "f1", "count", "accuracy"]
        results[count] = evaluate(model, X_test, y_test)
        count += 1

    mean = np.mean(results, axis = 0)
    std = np.std(results, axis = 0)
    print("##### Results #####")
    for m in range(0,len(metrics_names)):
        print(f'{metrics_names[m]} mean is: {mean[m]}, std. deviation is: {std[m]}')

    return results
```

```python
train = kfold_train(X, Y, n_splits = 10)
```

# BERT
We explore a simple BERT model, since Transformer based architectures have demonstrated good performance with text classification tasks. Because of hardware limitations I won't be training this across multiple runs or tuning it very precisely.

```python
df_dense = df_balanced.copy()
df_dense['text'] = df_dense['text'].apply(lambda text: clean_text(text, string = True))
df_dense.head()
```

We can split the data and one-hot encode the labels.

```python
from sklearn.model_selection import train_test_split
# train-test split
train, test = train_test_split(
    df_dense,
    train_size=0.90,
    stratify=df_dense['airline_sentiment'])

X_train = train["text"]
X_test = test["text"]
y_train = pd.get_dummies(train["airline_sentiment"], prefix = "airline_sentiment").to_numpy().astype('float64')
y_test = pd.get_dummies(test["airline_sentiment"], prefix = "airline_sentiment").to_numpy().astype('float64')
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import Model
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.layers import Layer
import keras.backend as K
import tensorflow_addons as tfa
```

## BERT's online implementation
Tensorflow provides a pre-trained BERT model for tuning. I use the smallest model due to the relative simplicity of the problem and hardware limitations. 

```python
tfhub_handle_encoder = \
    "https://tfhub.dev/tensorflow/small_bert/bert_en_uncased_L-2_H-128_A-2/2"
tfhub_handle_preprocess = \
    "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
```

```python
'''
This method constructs a model from Tensorflow hub, given the encoder and preprocessing handles.
Can be used with multiple text-processing models in Tensorflow hub.
It uses the Keras sequential model API for simplicity.
'''
def build_simpleBERT(tfhub_handle_encoder, tfhub_handle_preprocess):

    # ----- Defining model -----
    # defining layers
    text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')  
    preprocessing_layer = hub.KerasLayer(tfhub_handle_preprocess, name='preprocessing')
    encoder = hub.KerasLayer(tfhub_handle_encoder, trainable=True, name='BERT_encoder')
    
    # processing
    encoder_inputs = preprocessing_layer(text_input)
    outputs = encoder(encoder_inputs)
    x = outputs['pooled_output']
    x = tf.keras.layers.Dropout(0.2)(x)
    x = tf.keras.layers.Dense(
        3, activation='softmax', name='classifier')(x)
    
    model = tf.keras.Model(text_input, x)
    model.summary()
    return model
```

```python
model = build_simpleBERT(tfhub_handle_encoder, tfhub_handle_preprocess)
```

Categorical cross entropy as a loss function is used for optimization since this is a simple multi-class classification problem. Metrics for model evaluation, much like with logistic regression, are chosen to provide a holistic picture of how the model is learning. 

```python
loss = tf.keras.losses.CategoricalCrossentropy(from_logits = False)
metrics = [tf.metrics.CategoricalAccuracy('accuracy'), 
           tf.keras.metrics.AUC(name = 'AUC'),
           tf.keras.metrics.Precision(name ='precision'),
           tf.keras.metrics.Recall(name = 'recall'),
          ]
```

```python
# define training and validation parameters
batch_size = 64
epochs = 15
validation_split = 0.15
learning_rate = 1e-4


# define callbacks
earlyStopping = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 3)
decayLR = ReduceLROnPlateau(monitor = 'val_loss', factor = 0.1, patience = 1, verbose = 1)
callbacks = [decayLR, earlyStopping]


# compile model
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = learning_rate), 
      loss = loss,
      metrics = metrics
     )


history = model.fit(X_train, y_train,
                    batch_size = batch_size,
                    epochs = epochs,
                    callbacks = callbacks,
                    validation_split = validation_split,
                    shuffle = True)
```

```python
model.evaluate(X_test, y_test)
```

## Performance
The model performs a little more poorly compared to the logistic regression model, with accuracy, precision and recall about 2% lower at around 0.87. I also empirically observed heavy overfitting, likely due to BERT's relative complexity despite the small dataset.

A more complex model is probably better able to learn the noise introduced in the dataset from the semi-supervised approach. I did some simple regularization by introducing a dropout layer for the BERT model and introducing callbacks to reduce the learning rate and stop early.
